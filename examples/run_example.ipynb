# Example for running GlueShower in a Python Noteboook

```python
# import required functions from the accompanying Python Script

import glue_shower_functions as gsf
import time
```

```python
# example inputs: 

# gsf.file_output_showers(M0, m0, Nc, c, d, num_events, plasma_option)

# M0 = centre of mass energy (GeV)
# m0 = 0++ glueballs mass (GeV)
# Nc = number of colours in SU(Nc)
# c = Hadronisation scale multiplier, sets scale at c * (2 * m0)
# d = Temperature scale multiplier, N ~ e^(m/(d * Tc))
# num_events = number of events
# plasma_option = At hadronisation scale, gluon is turned into a plasma fireball and then decays
#                 (intstead of becoming single glueball)

gsf.file_output_showers(100, 10, 3, 1, 1, 100, False)
```

```python
# Scan over 8 benchmark points (as found in arXiv: ) for range of M0 values 

m0 = 10
Nc = 3
num_events = 15000

M0_range = [ 100, 200, 500, 1000 ]

for i in M0_range:
    start_time = time.time()
    gsf.file_output_showers(i, m0, Nc, 1, 1, num_events, False)
    gsf.file_output_showers(i, m0, Nc, 2, 1, num_events, False)
    gsf.file_output_showers(i, m0, Nc, 1, 2, num_events, False)
    gsf.file_output_showers(i, m0, Nc, 2, 2, num_events, False)
    gsf.file_output_showers(i, m0, Nc, 4, 1, num_events, True)
    gsf.file_output_showers(i, m0, Nc, 6, 1, num_events, True)
    gsf.file_output_showers(i, m0, Nc, 4, 2, num_events, True)
    gsf.file_output_showers(i, m0, Nc, 6, 2, num_events, True)
    print((time.time() - start_time)/60)
```
